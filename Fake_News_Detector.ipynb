```python
import re
```

```python

def wordopt(text):
  #to convert words into lowercase
  text=text.lower()
  #to remove URLs
  text = re.sub(r'http\S+', '', text)
  #to remove HTML tags
  text = re.sub(r'<.*?>', '', text)
  #to remove punctuation
  text = re.sub(r'[^\w\s]', '', text)
  #to remove digits
  text = re.sub(r'\d', '', text)
  #to remove newline characters
  text = re.sub(r'\n', '', text)

  return text
```

```python
news['text']=news['text'].apply(wordopt)
```

```python
X=news['text']
Y=news['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2)
```

```python


from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)
```

```python


from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

model = LogisticRegression()
model.fit(X_train_vect, Y_train)
y_pred = model.predict(X_test_vect)

print(classification_report(Y_test, y_pred))
print(confusion_matrix(Y_test, y_pred))
```

```python


from sklearn.tree import DecisionTreeClassifier

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

# Vectorize the text data
vectorizer = TfidfVectorizer()
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)

# Train the decision tree classifier model
model = DecisionTreeClassifier()
model.fit(X_train_vect, Y_train)

# Make predictions on the test data
y_pred = model.predict(X_test_vect)

# Evaluate the model
print(classification_report(Y_test, y_pred))
print(confusion_matrix(Y_test, y_pred))
```

```python


from sklearn.ensemble import RandomForestClassifier

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

# Vectorize the text data
vectorizer = TfidfVectorizer()
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)

# Train the random forest classifier model
model = RandomForestClassifier()
model.fit(X_train_vect, Y_train)

# Make predictions on the test data
y_pred = model.predict(X_test_vect)

# Evaluate the model
print(classification_report(Y_test, y_pred))
print(confusion_matrix(Y_test, y_pred))
```

```python


from sklearn.ensemble import GradientBoostingClassifier

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

# Vectorize the text data
vectorizer = TfidfVectorizer()
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)

# Train the gradient boosting classifier model
model = GradientBoostingClassifier()
model.fit(X_train_vect, Y_train)

# Make predictions on the test data
y_pred = model.predict(X_test_vect)

# Evaluate the model
print(classification_report(Y_test, y_pred))
print(confusion_matrix(Y_test, y_pred))
```

```python


def output_label(text):
  text = wordopt(text)
  vectorized_text = vectorizer.transform([text])
  prediction = model.predict(vectorized_text)
  if prediction == 0:
    return "Fake News"
  else:
    return "True News"
```

```python


def manual_testing():
  """
  This function allows you to manually test the model by inputting a news article.
  """

  # Get the news article from the user
  news_article = input("Enter a news article: ")

  # Preprocess the news article
  preprocessed_article = wordopt(news_article)

  # Vectorize the preprocessed article
  vectorized_article = vectorizer.transform([preprocessed_article])

  # Make a prediction on the vectorized article
  prediction = model.predict(vectorized_article)

  # Print the prediction
  if prediction == 0:
    print("Fake News")
  else:
    print("True News")

# Call the manual testing function
manual_testing()
```
